```python
def kk(listt, juftortoq):
    natija = []
    for tt in listt:
        if juftortoq.lower() == "juft" and tt % 2 == 0:
            natija.append(tt)
        elif juftortoq.lower() == "toq" and tt % 2 == 1:
            natija.append(tt)
    return natija
kk(list(range(100)), "juft")
```
